# Instalacion de paquetes de Julia

Julia cuenta con una gran cantidad de paquetes alojados en github ademas de poder usar bibliotecas de Python,R,Java,C++,C/Fortran

-Una vez identificado el paquete requerido (https://juliapackages.com) accedemos al administrador de paquetes agregando corchete derecho "]":
- add nombre_paquete

```julia
] add Pluto
```

```julia
Realizamos una prueba para verificar la instalacion con:
test nombre_paquete
(Se realiza dentro del modo de administrador de paquetes pkg>)
```

```julia
] test Pluto
```
